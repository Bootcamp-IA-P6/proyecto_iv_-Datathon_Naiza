### En este apartado vamos a importar las librerías que vamos a utilizar 
#### pandas = tratado de datos, numpy = operaciones matématicas, matplotlit = Gráficos, scikit learn = Módelo de regresión

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


df = pd.read_csv('all-states-history.csv')
df
```

### Creamos una nueva variable que se llama new_data y el valor que será ejecutado serán tres columnas, date, state, deathConfirmed


```python
new_data = df[['date', 'state', 'deathConfirmed']]
new_data
```

### Con la función rename --> renombro mis columnas que se guardan en una variable que se llama change 
### change.head () está función imprime las primero 5 filas, pero solo se muestran 4 porque los indices van desde cero


```python

change = new_data.rename(columns={
    'date': 'fecha',
    'state': 'estado',
    'deathConfirmed': 'muertes_confirmadas'
})
change.head()
```

### creamos una variable change_sin_nan y le indicamos que elimine todos los valores de mi nueva datframe change con la función dropna

```python
change_sin_nan = change.dropna()

print(change_sin_nan.head(10))
```

###  Seleccionamos la función de pd.to_datetime para cambiar el formato a fecha 

```python
 
change_sin_nan['fecha'] = pd.to_datetime(change_sin_nan['fecha'], format = 'mixed')
print(change_sin_nan)
```

### Esta línea es un filtro. Le estás pidiendo a Pandas que te muestre solo las filas donde hubo al menos 1 muerte.

```python
# Muertes confirmadas 
change_sin_nan[change_sin_nan['muertes_confirmadas']>0]
change_sin_nan.head(16)
```

### Con está función estamos diciendole a python que me filtre la primera fecha desde la primer muerte a la ultima 
#### 

```python
primer_año = change_sin_nan['fecha'].dt.year.max()
ultimo_año = change_sin_nan['fecha'].dt.year.min()
print(f"Las muertes empezaron en el año : {primer_año} \n y las ultimas fueron en el: {ultimo_año}")
```

### En el esatado AR haz una sumatorio total de la filas, solo imprime True si hay datos

```python
(change_sin_nan['estado'] == 'AR').sum()
```

### saber cuantos estados hay y de que tipo son, esta función me devuelve una lista ob type string

```python
#saber cuantos estados hay y de que tipo son 

change_sin_nan['estado'].unique()
```

### 2. Filtramos y contamos (esto te dará el conteo uno por uno)
### En total tenemos 64 estados 

```python
mis_estados = ['AL', 'AR', 'AZ', 'CO', 'CT', 'DE', 'GA', 'HI', 'ID', 'IL', 'IN', 'KY', 'LA',
 'MA', 'MD', 'ME', 'MI', 'MN', 'MP', 'MS', 'NC', 'NJ', 'OH', 'PR', 'SC', 'TN',
 'VA', 'VI', 'WI', 'WV', 'AK', 'VT', 'WA', 'FL', 'TX', 'NV', 'NY']


conteo = change_sin_nan[change_sin_nan['estado'].isin(mis_estados)]['estado'].value_counts()
print(conteo.head(20))

```

### Con esta función sabemos la cantidad de muertes confirmdas que hubó por estado 

```python

muertes_totales = change_sin_nan[change_sin_nan['estado'].isin(mis_estados)].groupby('estado')['muertes_confirmadas'].sum()
print(muertes_totales)
```

```python

# --- PASO DE SEGURIDAD 2: Filtrar y limpiar ---
# Nos aseguramos de que no haya ceros o vacíos que rompan la lógica
df_al = change_sin_nan[change_sin_nan['estado'] == 'AL'].copy()

df_al = df_al.sort_values('fecha')

# 1. Convertimos la fecha a número de días
df_al['dias'] = (df_al['fecha'] - df_al['fecha'].min()).dt.days

# 2. Definimos X (entrada) y y (salida)
X = df_al[['dias']] 
y = df_al['muertes_confirmadas']

# 3. Creamos, entrenamos y predecimos
modelo = LinearRegression()
modelo.fit(X, y)
prediccion = modelo.predict(X)

# 4. Graficamos
plt.figure(figsize=(10, 6))
plt.scatter(df_al['dias'], y, color='blue', label='Datos reales', s=10)
plt.plot(df_al['dias'], prediccion, color='red', linewidth=2, label='Tendencia Lineal')
plt.title('Regresión Lineal: Tendencia de Muertes en Alabama')
plt.xlabel('Días transcurridos')
plt.ylabel('Muertes Confirmadas')
plt.legend()
plt.show()
```
